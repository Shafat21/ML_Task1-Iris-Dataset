```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')

df.head()
```

```python
# Display basic information about the dataset
df.info()

# Check the column names (features)
df.columns
```

```python
# If the target values are already present as species names, skip this step
# If numeric target values are present, map them to species names
species_names = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}
if 'target' in df.columns:
    df['species'] = df['target'].map(species_names)
    
# Display the updated DataFrame
df.head()
```

```python
# Scatter plot for Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

# Scatter plot for Petal Length vs Petal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Set up the matplotlib figure for multiple plots
plt.figure(figsize=(12, 10))

# Sepal Length Boxplot
plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=df, hue='Species', palette='Set2', dodge=False)
plt.title('Sepal Length')

# Sepal Width Boxplot
plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=df, hue='Species', palette='Set2', dodge=False)
plt.title('Sepal Width')

# Petal Length Boxplot
plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLengthCm', data=df, hue='Species', palette='Set2', dodge=False)
plt.title('Petal Length')

# Petal Width Boxplot
plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidthCm', data=df, hue='Species', palette='Set2', dodge=False)
plt.title('Petal Width')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the mean of each feature grouped by species
df_grouped = df.groupby('Species').mean()

# Plot the data as a grouped bar chart
df_grouped.plot(kind='bar', figsize=(10, 6), color=['#cc5531', '#31bccc', '#99FF99', '#31cc74'])
plt.title('Average Measurements for Each Species')
plt.ylabel('Average Value')
plt.xticks(rotation=0)
plt.legend(title='Features')
plt.show()
```

```python
# Scatter plot with regression lines for Sepal Length vs Sepal Width
sns.lmplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df, palette=['orange', 'blue', 'maroon'], markers=['o', 's', 'D'], height=6)
plt.title('Sepal Length vs Sepal Width with Regression Lines')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
